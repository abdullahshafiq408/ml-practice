### What's in the Notebook?
I practiced a few common feature engineering techniques which I am sharing here in this notebook.

```python
# imports
import pandas as pd
import numpy as np
import sklearn
```

```python
# One-Hot Encoding
data= {'color': ['Red', 'Green', 'Velvet', 'Brown', 'Green', 'Red', 'Brown', 'Brown']}
```

```python
df= pd.DataFrame(data)

df
```

```python
# the actual one hot encoding step
df_encoded= pd.get_dummies(df, columns= ['color'])
df_encoded
```

```python
# Binning
data = {'Age': [23, 45, 18, 34, 67, 50, 21]}
df = pd.DataFrame(data)
```

```python
df
```

```python
# let's make bins
bins= [0, 20, 40, 60, 100]
labels= ['0-20', '21-40', '41-60', '61-100']
df['Age-Group']= pd.cut(df['Age'], bins= bins, labels=labels)
df
```

```python
# Feature Splitting
data = {'Full_Address': [
    '123 Elm St, Springfield, 12345', '456 Oak Rd, Shelbyville, 67890']}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Using regular expressions
df[['Street', 'City', 'Zipcode']] = df['Full_Address'].str.extract(
    r'([0-9]+\s[\w\s]+),\s([\w\s]+),\s(\d+)')
```
