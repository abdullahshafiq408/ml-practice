### What's in the notebook?
I generated a random dataset and then trained two different models on it to get results from GridSearchCV and RandomizedSearchCV.

#### As you can see I trained the tree_cv twice and each time it came out with different results and even different score. 
#### It is because it randomly picks the parameters to test on and hence is not optimal.
#### Yet, it is more useful than GridSearchCV as it is time effective.
#### For a more optimal and useful soltuion we can use Bayesian Optimization which we'll discuss in another notebook.

```python
# GridSearchCV
```

```python
# importing important libraries
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.datasets import make_classification
```

```python
# generating a random dataset
```

```python
X, y= make_classification(n_samples= 1000, n_features= 20, n_classes= 2, n_informative= 10, random_state= 42)
```

```python
X
```

```python
# for value of C in regularization
c_space= np.logspace(-5, 8, 15)
```

```python
c_space
```

```python
# defining the parameters we are going to test
param_grid= {
    'C': c_space,
    'penalty': ['l1', 'l2']
}
```

```python
# initializing the logistic regression model
logreg= LogisticRegression(solver= 'liblinear')
logreg_cv= GridSearchCV(logreg, param_grid, cv= 5)
```

```python
logreg_cv.fit(X, y)
```

```python
# checking best parameters for us out of the pre-defined
print(f'Tuned parameters: {logreg_cv.best_params_}')
print(f'Best score: {logreg_cv.best_score_}')
```

```python
# Randomized Search CV
```

```python
# we'll need a few more libraries
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# defining the parameters to test on
random_params= {
    'max_depth': [3, None],
    'max_features': randint(1, 9),
    'min_samples_leaf': randint(1,9),
    'criterion': ['gini', 'entropy']
}
```

```python
# initializing and training the model
tree= DecisionTreeClassifier()
tree_cv= RandomizedSearchCV(tree, random_params, cv= 5)
tree_cv.fit(X, y)
```

```python
# looking for the best parameters. Attempt 1
print(f'Tuned parameters: {tree_cv.best_params_}')
print(f'Best score: {tree_cv.best_score_}')
```

```python
tree= DecisionTreeClassifier()
tree_cv= RandomizedSearchCV(tree, random_params, cv= 5)
tree_cv.fit(X, y)
```

```python
# looking for the best parameters. Attempt 2
print(f'Tuned parameters: {tree_cv.best_params_}')
print(f'Best score: {tree_cv.best_score_}')
```
