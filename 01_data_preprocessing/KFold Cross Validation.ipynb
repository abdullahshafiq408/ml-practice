### What's in the notebook?
I tried KFold Cross Validation on Iris Dataset and also checked the cross val score. This practice notebook helped me getting the idea of different sort of cross validations, their required parameters and the score checking.


```python
# importing important libraries
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
X, y= iris.data, iris.target
```

```python
X
```

```python
svm_classifier= SVC(kernel= 'linear')
```

```python
# intializing our kfold object
```

```python
num_folds= 5
kf= KFold(n_splits= num_folds, shuffle= True, random_state= 42)
```

```python
# cross val score performs validation across all folds itself
```

```python
cross_val_results= cross_val_score(svm_classifier, X, y, cv= kf)
```

```python
cross_val_results
```

```python
# representing the scores of different folds
```

```python
for i, result in enumerate(cross_val_results):
    print(f'Fold {i}: {round(result*100, 2)}%')
```
