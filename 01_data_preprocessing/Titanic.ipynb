## What's in the notebook?
I cleaned dataset called 'titanic' by:-
-> removing duplicates
-> handling missing values
-> analyzing and removing outliers 
-> scaling(normalizing) the dataset.

### How to have this dataset?
It is publicly available on Kaggle.com with many variants. I used a messy one.

```python
# importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# reading dataset
df= pd.read_csv('Titanic-Dataset.csv')

# analyzing dataset manually
df.head()

df.shape

df.info()

# looking for duplicate values
df.duplicated()

# making lists for categorical columns and nmerical columns as they'll later help us
cat_col= [col for col in df.columns if df[col].dtype== 'object']
num_col= [col for col in df.columns if df[col].dtype!= 'object']

cat_col

num_col

# figuring out null values
df[cat_col].nunique()

df[num_col].nunique()

round((df.isnull().sum()/ df.shape[0])*100, 2)

df.head()

# dropping columns which don't add much value to our dataset
df1= df.drop(columns= ['Name', 'Ticket', 'Cabin'])

df1

# dropping null-heavy columns and handling missing values
df1.dropna(subset=['Embarked'], inplace= True)

df1['Age']= df1['Age'].fillna(df1['Age'].mean())

round((df1.isnull().sum()/ df1.shape[0])*100, 2)

# plotting boxplot to check outliers
plt.boxplot(df1['Age'], vert= False)
plt.ylabel('Variable')
plt.xlabel('Age')
plt.title('BoxPlot')
plt.show()

# taking only 95% of the most representative values (excluding outliers)
mean= df1['Age'].mean()
std= df1['Age'].std()

mean

std

lower_bound= mean- 2*std
upper_bound= mean+ 2*std


lower_bound

upper_bound

df2= df1[(df1['Age']>=lower_bound) & (df1['Age']<=upper_bound)]

df2

df3= df2.fillna(df['Age'].mean())
df3.isnull().sum()

# separating input and output features
X= df3[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Age', 'Fare', 'Embarked']]

cat_col

num_col

X

Y= df3[['Survived']]

Y

# Scaling using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

num_col_= [col for col in X.columns if df[col].dtype!= 'object']

x1= X
x1[num_col_]= scaler.fit_transform(x1[num_col_])

x1
```
