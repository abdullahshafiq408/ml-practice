I implemented gradient descent using numpy and then using it I trained my linear regression model on a sample dataset, analyzing the results and convergence.

```python
# Simple Linear Regression
```

```python
# importing libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# generating a dataset
X, y= make_regression(n_samples= 100, n_features= 1, noise= 15, random_state= 42)
y
```

```python
# reshaping the list into a column vector
y= y.reshape(-1, 1)
y
```

```python
m= X.shape[0]
m
```

```python
X_b= np.c_[np.ones((m,1)), X]
```

```python
X_b
```

```python
theta= np.array([[2.0], [3.0]])
```

```python
theta
```

```python
plt.figure(figsize= (6, 4))
plt.scatter(X, y, color='blue', label= 'actual data')
plt.plot(X, X_b.dot(theta), color="green", label="Initial Line (No GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
```

```python
# Gradient Descent
```

```python
learning_rate= 0.01
epochs= 100

for _ in range(epochs):
    y_pred= X_b.dot(theta)
    gradients = (2 / m) * X_b.T.dot(y_pred - y)
    theta-= (learning_rate*gradients)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```
