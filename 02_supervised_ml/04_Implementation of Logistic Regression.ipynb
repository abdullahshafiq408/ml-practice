#### Logistic Regression implementation using Gradient Descent
I implemented linear regression using numpy and then analyzed the results on a sample dataset.

```python
# importing important libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
```

```python
# simple squisher
def sigmoid(z):
    z= np.clip(z, -250, 250)
    return 1/(1+np.exp(-z))
```

```python
# core training logic
def train_logistic_regression(X, y, learning_rate= 0.1, iterations= 1000):
    m, n= X.shape
    X_b= np.c_[np.ones((m,1)), X]
    weights= np.zeros(n+1)

    for _ in range(iterations):
        linear_model= np.dot(X_b, weights)
        y_predicted= sigmoid (linear_model)
        gradients = (1 / m) * np.dot(X_b.T, (y_predicted - y))
        weights -= learning_rate * gradients
    return weights
```

```python
# simple predictions using the weights given by train_logistic_regression
def predict(X, weights, threshold=0.5):
    m = X.shape[0]
    X_b = np.c_[np.ones((m, 1)), X]
    probabilities = sigmoid(np.dot(X_b, weights))
    
    return [1 if p >= threshold else 0 for p in probabilities]
```

```python
# dataset
X, y= make_classification(n_samples=1000, n_classes= 2, n_features= 4, random_state=42)
```

```python
X
```

```python
w= train_logistic_regression(X, y)
```

```python
y_pred= predict(X, w)
```

```python
# final accuracy
acc = accuracy_score(y, y_pred) * 100
print(f"Logistic Regression model accuracy: {acc:.2f}%")
```
